```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0,1,1000)
```

```python
width = 0.01
histmin = np.floor(min(x))
histmax = np.ceil(max(x))+width

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=bins,alpha=0.5,edgecolor="olivedrab",color="yellowgreen")
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("journal-histo.png",bbox_inches="tight")
```
