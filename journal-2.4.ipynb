```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.triangular(0, 0.3, 1, 1000)
```

```python
width = 0.01
histmin = np.floor(min(x))
histmax = np.ceil(max(x))+width

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=bins,alpha=0.5,edgecolor="palevioletred",color="mistyrose")
plt.ylabel=("N per bin")
plt.xlabel=("x")
plt.savefig("journal-histo2.png",bbox_inches="tight")
```
